```python
import requests
from bs4 import BeautifulSoup
from prettytable import PrettyTable

def scrape_books(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extracting data from each book listing
        books = soup.find_all('article', class_='product_pod')

        # Creating a PrettyTable
        table = PrettyTable()
        table.field_names = ["Title", "Rating", "Price", "Link"]

        # Looping through each book and adding data to the table
        for book in books:
            title = book.h3.a['title']
            rating = book.p['class'][1]  # Assuming the second class represents the rating
            price = book.select('div p.price_color')[0].text
            link = url + book.h3.a['href']

            # Adding a row to the table
            table.add_row([title, rating, price, link])

        # Printing the table
        print(table)

        # Check for the next page and scrape its data
        next_page = soup.find('li', class_='next')
        if next_page:
            next_page_url = url.rsplit('/', 2)[0] + '/' + next_page.a['href']
            scrape_books(next_page_url)

    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

# Starting URL for the first page
start_url = 'http://books.toscrape.com/'

# Scrape books from all pages
scrape_books(start_url)
```
